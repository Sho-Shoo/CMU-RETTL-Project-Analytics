Here we explore what values each column can take: 

```python
maxLength = 10
for colName in tutorLogDF.columns: 
    col = tutorLogDF[colName]
    print("For column *", colName, "* the unique values are: ")
    if(len(pd.unique(col)) < maxLength): 
        print(pd.unique(col))
    else: 
        # if length is too long, going to omit some values 
        print(pd.unique(col)[0:maxLength])
        print(" ... ")
    print()
```
